```python
import pandas as  pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("movies.csv")
```

```python
def statics_overview(df):
    print("\n top ten rows \n", df.head(10))
    print("\n sample as rows \n", df.sample(10))
    print("\n last rows \n", df.tail())
    print("\n size of dataset \n", df.shape)
    print("\n types of data sets \n", df.dtypes)
    print("\n print info \n", df.info())
    return
statics_overview(df)
```

```python

print("Missing values")
for col in df.columns:
    m_values=df[col].isna().sum()
    percentage=(m_values/df.shape[0])*100
    print("%s: %.2f%% (%d)" % (col,percentage,m_values))
```

```python
df.rename(columns=str.lower, inplace=True)
```

```python
df[['stars', 'genre', 'one-line']]=df[['stars', 'genre', 'one-line']].replace(r'[^a-zA-Z0-9]',"")

df['genre']=df['genre'].str.replace('/n','').str.strip()
df['stars']=df['stars'].str.replace(r'\n', ' ', regex=True).str.strip()
df['one-line']=df['one-line'].str.replace('/n','').str.strip()
df[['stars','genre','one-line']]
```

```python
df['stars']
```

```python
def extract_info(s):
    s=s.replace('..','').strip()
    info={'Director':"",'Stars':''}

    parts=[p.strip() for p in s.split('|')]
    for i in parts:
        if i.startswith('Director:') or i.startswith('Directors:'):
            info['Director']=i.split(":",1)[1].strip()
        elif i.startswith('Star:')or i.startswith('Stars:'):
            info['Stars']=i.split(':',1)[1].strip()
    return pd.Series(info)
df[['Director', 'Stars']] = df['stars'].apply(extract_info)
df[['stars', 'Stars', 'Director']]
    
```

```python
df.drop("stars",axis=1, inplace=True)
```

```python
df['year']=df['year'].astype(str)
df['year']=df['year'].str.extract(r'((?:19|20)\d{2})')
df['year'] = pd.to_datetime(df['year'], format="%Y", errors='coerce').dt.year.astype('Int64')
df['year']
```

```python
def remove_dup(df):
    print("\n number of duplicate \n", df.duplicated().sum())
    print(f"\n remove duplicated is {df.drop_duplicates(inplace=True)}\n")
    return
remove_dup(df)
```

```python
df['votes']=df['votes'].astype(str).str.replace(',','').str.strip()
df['votes']=pd.to_numeric(df['votes'], errors='coerce')
def missing_v(df):
    print(f"votes total number of nan {df['votes'].isnull().sum()}\n")
    print(df['votes'].fillna(0, inplace=True))
    print(f"\n after fillna total number of nan {df['votes'].isnull().sum()}\n")
    return
missing_v(df)
```

```python
df['Director']=df['Director'].replace('',np.nan).str.strip()
df['Director']=df['Director'].fillna('Not known')
df['Director']=df['Director'].astype(str)
```

```python
df['gross']=df['gross'].str.replace('$','').str.strip()
```

```python
def clean_val(val):
    if isinstance(val,str):
        val=val.strip().upper()
        if 'M' in val:
            return float(val.replace('M',''))*1000000
        elif 'K' in val:
            return float(val.replace('K',''))*1000
        elif 'B' in val:
            return float(val.replace('B',''))*1_000_000_000
        else:
            try:
                return float(val.replace(',',''))
            except:
                return None
    return None
df['gross']=df['gross'].apply(clean_val)

            
                         
            
```

```python
df['genre'].isna().sum()
```

```python
df['genre']=df['genre'].astype(str).str.replace('\n', '', regex=True).str.strip().str.replace(r'\s*,\s*', ', ', regex=True)
```

```python
merger_df=df.groupby('Director')['genre'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else'Unknow')
df['genre']=df['genre'].fillna(df['Director'].map(merger_df))
df['genre'] = df['genre'].fillna('Unknown')  
df['genre'].isna().sum()
                               
```

```python
df['one-line']=df['one-line'].str.replace('Add a Plot','Unknown')
```

```python
# Fill NaN based on group median (Director + GENRE)
df['rating']=df.groupby(['genre','Director'])['rating'].transform(lambda x: x.fillna(x.median()))
```

```python
median=df['rating'].median()
df['rating'].fillna(median, inplace=True)
```

```python
df['rating']
```

```python
df['runtime']=df.groupby(df['genre'].str.replace(',','',regex=True).str.strip())['runtime'].transform(lambda x:x.fillna(x.median()))
median=df['runtime'].median()
df['runtime'].fillna(median,inplace=True)
```

```python
Q1=df['runtime'].quantile(0.25)
Q3=df['runtime'].quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR

df=df[(df['runtime']>=lower) &(df['runtime']<=upper)]
```
